```python
sns.relplot(x='speeding', y='alcohol', data=df, hue='not_distracted')
```

```python
plt.subplots(figsize=(20,15))
sns.barplot(y='abbrev', x='alcohol', data=df)
```

```python
plt.subplots(figsize=(20,15))
sns.barplot(x='total', y='speeding', data=df)
```

inference:the barplot displays the mean total crashes for different levels of alcohol consumption here the alcohol consumption is direcly proportional to total crashes

```python
sns.jointplot(x='total', y='speeding', data=df)
```

inference:from the joint plot it can be observed that there is no specific patterns in speeding and total
